```python
import librosa
import numpy as np
import pandas as pd
import math
import soundfile as sf
import IPython.display as ipd
```

```python
Synth, sr_Synth = librosa.load("audio/Synth.wav", sr = 16000)
spectrogram_Synth = librosa.stft(Synth, n_fft=2048, hop_length=256, center=False, win_length=1024) 
M_Synth = abs(spectrogram_Synth)
phase_Synth = spectrogram_Synth/(M_Synth + 2.2204e-16) 
Piano, sr_Piano = librosa.load("audio/Piano.wav", sr = 16000)
spectrogram_Piano = librosa.stft(Piano, n_fft=2048, hop_length=256, center=False, win_length=1024) 
M_Piano = abs(spectrogram_Piano)
phase_Piano = spectrogram_Piano/(M_Piano + 2.2204e-16) 
Blinding, sr_Blinding = librosa.load("audio/BlindingLights.wav", sr = 16000)
spectrogram_Blinding = librosa.stft(Blinding, n_fft=2048, hop_length=256, center=False, win_length=1024) 
M_Blinding = abs(spectrogram_Blinding)
phase_Blinding = spectrogram_Blinding/(M_Blinding + 2.2204e-16)
```

```python
T=np.dot(M_Piano,np.linalg.pinv(M_Synth))
```

```python
newDataset = np.dot(T,M_Synth)-M_Piano
finalDataset = np.linalg.norm(newDataset, 'fro')**2
print(finalDataset)
```

```python
df = pd.DataFrame(T, columns=None)  


df.to_csv('problem3t.csv', index=False)  

print("Array saved to 'problem3t.csv'")
```

```python
MD = T @ M_Blinding 
MD_spectogram = MD * np.exp(1j*phase_Blinding) 

YD = librosa.istft(MD_spectogram, hop_length=256, win_length=1024, center=False) 
print(YD.shape)

sf.write('problem3.wav', YD, 16000)
ipd.Audio('problem3.wav', rate=16000)
```
